#Filtrado de Datos
Filtra el DataFrame para mostrar solo las filas donde las ventas (Ventas) sean mayores a 50.

Filtra el DataFrame para mostrar solo las filas donde el precio (Precio) sea menor a 0.5.

Utilizando el método query(), filtra el DataFrame para mostrar las filas donde el producto sea Manzanas y las ventas sean mayores a 30.

```python
# Filtrado de datos
print("Ventas mayores a 50:")

ventas_mayores_50 = df[df["Total Amount"] > 50]
ventas_mayores_50
```

```python
print("Precios menores a 0.5:")

precio_menor_05 = df[df["Price per Unit"] < 0.5]
precio_menor_05
```

```python
# verificando que no hay precios menores a 0.5
(df["Price per Unit"] < 0.5).sum()
```

```python
# Utilizando el método query(), filtra el DataFrame para mostrar las filas donde el producto sea Manzanas y las ventas sean mayores a 30. Manzanas no existe en el archivo, por ende reemplazo por beauty, y ventas por la cantidad total.
print("Filas con producto Beauty y cantidad total mayores a 30:")

beauty_sales = df[(df["Product Category"] == "Beauty") & (df["Total Amount"] > 30)]
beauty_sales
```
